```python
#This function takes in a revids and my headers (constant) and creates a json object
import requests
import json

headers = {'User-Agent' : 'https://github.com/abhiv91', 'From' : 'abhiv@uw.edu'}

def get_ores_data(revision_ids, headers):
    
    # Define the endpoint
    endpoint = 'https://ores.wikimedia.org/v3/scores/{project}/?models={model}&revids={revids}'
    
    # Specify the parameters - smushing all the revision IDs together separated by | marks.
    # Yes, 'smush' is a technical term, trust me I'm a scientist.
    # What do you mean "but people trusting scientists regularly goes horribly wrong" who taught you tha- oh.  
    params = {'project' : 'enwiki',
              'model'   : 'wp10',
              'revids'  : '|'.join(str(x) for x in revision_ids)
              }
    api_call = requests.get(endpoint.format(**params))
    response = api_call.json()
    return response


# So if we grab some example revision IDs and turn them into a list and then call get_ores_data...
example_ids = [783381498, 807355596, 757539710]
example = get_ores_data(example_ids, headers)
```

```python
#We can iterate through revids, calling our json function 100 revids at a time
#we store the revids and their scores on a dictionary
#We'll skip any records that show a key error as that revid maynot be available from the api
example_dict={}
for y in range(472):
    try:
        print (y)
        example = get_ores_data(revids[y], headers)
    
        
        for x in revids[y]:
    
            example_dict[x]= example['enwiki']['scores'][(str(x))]['wp10']['score']['prediction']
    except KeyError:
        
        continue
example_dict
```

```python
#saving as csv as backup
with open('revidscored.csv', 'w') as f:  # Just use 'w' mode in 3.x
    w = csv.DictWriter(f, example_dict.keys())
    w.writeheader()
    w.writerow(example_dict)
```

```python
#we want to get rid of the commas in the population so we can cast to int.
#we need to cast to int since we will need to divide with these numbers
popheaders = ['Country','Population']
df_pop=pd.DataFrame(population,columns=popheaders)
intlist=[]
for a in df_pop['Population']:
    a=int(a.replace(',', ''))
    intlist.append(a)
```
